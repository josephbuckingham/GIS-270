```python
import geopandas as gpd
```

```python
fire_planted = gpd.read_file('../../processing/fire_planted/fire_planted.gdb')
harvest_planted = gpd.read_file('../../processing/harvest_planted/harvest_planted.gdb')
fire_planted.plot()
harvest_planted.plot()
```

```python
fire_planted
```

```python
harvest_planted
```

```python

harvest_planted = harvest_planted.rename({'raster_val': 'harvest_year'}, axis=1)
fire_planted = fire_planted.rename({'raster_val': 'fire_year'}, axis=1)
fire_planted
```

```python
harvest_planted[['ACTIVITY_TREATMENT_UNIT_ID', 'harvest_year']]
```

```python
matching_geometry = fire_planted.merge(harvest_planted[['ACTIVITY_TREATMENT_UNIT_ID', 'harvest_year']], on='ACTIVITY_TREATMENT_UNIT_ID', how='inner')
matching_geometry
```

```python
print(matching_geometry[['ATU_COMPLETION_DATE', 'fire_year', 'harvest_year']])
```

```python
year_formatted_matching = matching_geometry.copy()
year_formatted_matching['ATU_COMPLETION_DATE'] = matching_geometry['ATU_COMPLETION_DATE'].dt.year
# matching_geometry['ATU_COMPLETION_DATE'].dt.year
print(year_formatted_matching[['ATU_COMPLETION_DATE', 'fire_year', 'harvest_year']])
```

```python
year_formatted_matching[(year_formatted_matching['harvest_year']) < (year_formatted_matching['fire_year'])][['ATU_COMPLETION_DATE', 'fire_year', 'harvest_year' ]]
```

```python
year_formatted_matching[(year_formatted_matching['fire_year'] <= year_formatted_matching['ATU_COMPLETION_DATE'])][['ATU_COMPLETION_DATE', 'fire_year', 'harvest_year' ]]
```

```python
matching_given_to_fire = year_formatted_matching[(year_formatted_matching['fire_year'] <= year_formatted_matching['ATU_COMPLETION_DATE'])]
```

```python
matching_given_to_fire = matching_given_to_fire.drop('harvest_year', axis=1)
```

```python
matching_given_to_fire = matching_given_to_fire.rename({'fire_year': 'raster_val'}, axis=1)
matching_given_to_fire
```
